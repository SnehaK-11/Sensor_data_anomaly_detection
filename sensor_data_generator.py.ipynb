```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
engine = np.random.default_rng(11)
points = 1000
start = datetime.now()
time_series = []
for t in range(points):
    time_series.append(start + timedelta(seconds=t))
clean_signal = np.sin(
    np.linspace(0, 25, points)
) * 8 + 48
shake = engine.normal(0, 1.8, points)
stream = clean_signal + shake
labels = np.zeros(points)
places = engine.choice(points, 50, replace=False)
for p in places:
    mode = engine.integers(0, 3)
    if mode == 0:
        jump = engine.integers(22, 38)
        stream[p] = stream[p] + jump
    elif mode == 1:
        fall = engine.integers(22, 38)
        stream[p] = stream[p] - fall
    else:
        mix = engine.integers(-28, 28)
        stream[p] = stream[p] + mix
    labels[p] = 1
table = pd.DataFrame({
    "timestamp": time_series,
    "sensor_value": stream,
    "is_anomaly": labels
})
table.to_csv("my_sensor_stream.csv", index=False)
print("my_sensor_stream.csv created.")
```
